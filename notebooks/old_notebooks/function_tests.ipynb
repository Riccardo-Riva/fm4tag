```python
import numpy as np
from numpy.lib.recfunctions import structured_to_unstructured as s2u
```

```python
help(s2u)
```

```python
structured_dtype = np.dtype(
    [('a', 'f4'), ('b', 'i4'), ('c', 'f8'), ('d', 'i1'), ('e', '?')]
)
```

```python
array = np.ones((3,), dtype=structured_dtype)
```

```python
array
```

```python
s2u(array, dtype=float)
```

```python
array_2 = array[['a', 'b', 'c']]
```

```python
s2u(array_2)
```
